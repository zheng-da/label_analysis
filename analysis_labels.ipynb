```python
import dgl
import torch as th
from ogb.nodeproppred import DglNodePropPredDataset
```

```python
dataset = DglNodePropPredDataset(name = 'ogbn-products')

split_idx = dataset.get_idx_split()
train_idx, valid_idx, test_idx = split_idx["train"], split_idx["valid"], split_idx["test"]
graph, label = dataset[0]
```

```python
uniq_label, cnt = th.unique(label, return_counts=True)
print(uniq_label)
print(cnt)
```

```python
for i, l in enumerate(uniq_label):
    print('label:', l.item())
    print('#nodes:', cnt[i].item())
    nid = th.nonzero(label == l, as_tuple=True)[0]
    src, _ = graph.in_edges(nid)
    src_nid = th.unique(src)
    ls = label[src_nid]
    print('in-edges:', len(ls), th.sum(ls == l).item())
    
    _, dst = graph.out_edges(nid)
    dst_nid = th.unique(dst)
    ls = label[dst_nid]
    print('out-edges:', len(ls), th.sum(ls == l).item())
```
